```python
#IMPORTING LIBRARIES 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#READING DATA FROM CSV FILE
df = pd.read_csv('Flipkart_Mobiles.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#CHECKING NULL VALUES IN THE FILE
df.isnull().sum()
```

```python
df.shape
```

```python
#DROPING NULL VALUES
df.dropna(inplace = True)
```

```python
df.head()
```

```python
#NOW THE DATA DO NOT HAVE THE NULL VALUES
df.isnull().sum()
```

```python
# CREATING A DISCOUNT PERCENT COLUMN IN THE FILE FOR BETTER UNDERSTANDING
df['discount']=round((df['Original Price']-df['Selling Price'])/df['Selling Price']*100,2)
```

```python
df.head()
```

```python
df.sort_values(by= 'Brand' ,inplace = False, ascending = True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
df.describe()
```

```python
sorted_categories = df['Brand'].value_counts().sort_values(ascending=False).index
ax=sns.countplot(x='Brand',data=df,order=sorted_categories)
sns.set(rc= {'figure.figsize': (21,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
grouped = df.groupby('Brand')['Rating'].mean().sort_values(ascending = False)
plt.stem(grouped.index, grouped.values)
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.show()

```

```python
df_sorted = df.sort_values('discount', ascending=False)
sns.set(rc= {'figure.figsize': (20,8)})
sns.catplot(x='Brand',y='discount',kind='bar',data=df_sorted,height =8,aspect=2.5)
```

```python
df_sorted = df.sort_values('Selling Price', ascending=False)
sns.set(rc= {'figure.figsize': (20,8)})
sns.boxplot(x='Brand',y='Selling Price',data=df_sorted)
```

```python
df_sorted = df.sort_values('Original Price', ascending=False)
sns.set(rc= {'figure.figsize': (20,8)})
sns.boxplot(x='Brand',y='Original Price',data=df_sorted)
```

```python
sns.displot(df,x='Selling Price',bins = 5 , hue='Brand',aspect =2)
```

```python
top_rated_phones = df.sort_values(by='Rating', ascending=False).head(10)


top_rated_phones['Brand_Model'] = top_rated_phones['Brand'] + ' ' + top_rated_phones['Model']


plt.figure(figsize=(10,4))
sns.barplot(x='Brand_Model', y='Rating', data=top_rated_phones)
plt.xticks(rotation=90)
plt.title('Top rated phones', size=15)
plt.show()
```
